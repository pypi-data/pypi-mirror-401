```python
import minillmlib as mll
import os
from dotenv import load_dotenv
load_dotenv()

node = mll.ChatNode(content="Follow the audio instructions.", 
audio_data=mll.AudioData(audio_paths=[
  "./data/test.mp3"
]), image_data=mll.ImageData(images=[
  "./data/test.jpg"
]))

gi = mll.GeneratorInfo(
  model="google/gemini-2.0-flash-lite-001",
  api_url="https://openrouter.ai/api/v1/chat/completions",
  api_key=os.getenv("OPENROUTER_API_KEY"),  # Set your OpenRouter API key
  _format="url",  # Use URL format for OpenRouter
  completion_parameters=mll.GeneratorCompletionParameters(
    temperature=0.7,
    max_tokens=1024,
  )
)

result = node.complete_one(mll.NodeCompletionParameters(
  gi=gi,
  exp_back_off=False,
  retry=1
))
print(result.content)
```

```python
node = mll.ChatNode(
  content="Describe this video in great details, and transcribe the audio.",
  video_data=mll.VideoData(
    videos=["./data/test.mp4"]
  )
)

result = node.complete_one(mll.NodeCompletionParameters(
  gi=gi,
  exp_back_off=False,
  retry=1,
))
print(result.content)
```
