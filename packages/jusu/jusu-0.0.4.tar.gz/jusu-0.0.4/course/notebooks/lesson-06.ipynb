```python
from JUSU import Div, Button, Img, StyleSheet

styles = StyleSheet()
styles.add_class('box', {'width':'150px','height':'150px','background':'#ddd'})
styles.add_class('hidden', {'display':'none'})

page = Div(Button('Toggle', onclick="toggleClass(document.querySelector('.box'), 'hidden')"), Img(src='https://via.placeholder.com/150', cls='box'), cls='container')
page.register_css(styles)
page.render_to_file('examples/interactive.html', styles=styles)
print('Wrote examples/interactive.html')
```
