```python
from JUSU import Div, H1, P, StyleSheet

def card(title, body):
    s = StyleSheet()
    s.add_class("card", {"border": "1px solid #ddd", "padding": "1rem"})
    return Div(H1(title), P(body), cls="card", styles=s)

c = card("My Card", "Card body text")
c.bundle("examples", name="card-demo", scoped=True)
print('Wrote bundle in examples/')
```
