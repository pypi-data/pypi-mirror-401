```python
# Example: register a tiny autoprefixer that prefixes 'display: flex'
from JUSU.css import StyleSheet

def simple_prefix(styles):
    # naive: replace 'display: flex' with prefixed variants
    return styles.replace('display: flex', '-webkit-display: flex; display: flex')

s = StyleSheet()
s.add_autoprefixer(simple_prefix)
s.add_class('f', {'display':'flex'})
print(s.render())
```
