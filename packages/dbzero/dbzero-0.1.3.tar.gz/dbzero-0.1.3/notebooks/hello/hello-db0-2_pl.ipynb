```python
import dbzero as db0
db0.init(dbzero_root = "/dbzero", prefix="data")
```

```python
@db0.memo
class HelloWorld:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        print(f"Witaj {self.name}. Witamy w dbzero!!")
```

        <strong>Otw√≥rzmy instancjƒô:&nbsp;</strong> U≈ºyjemy wcze≈õniej wygenerowanego UUID (< wklej swoje UUID poni≈ºej >)
    </p>
</div>

```python
h = db0.fetch(HelloWorld, "HOHKGLN4SOLAXIEBUCAIAAIN")
```

```python
h.greet()
```

<div style="background-color: #f8f9fa; border: 2px solid #34495e; border-radius: 8px; padding: 20px; margin: 15px 0;">
    <div style="display: flex; align-items: flex-start; gap: 15px;">
        <div style="font-size: 2em; color: #34495e;">üì¶</div>
        <div>
            <p style="margin: 0; font-size: 1.1em; line-height: 1.6; color: #2c3e50;">
                JednƒÖ z mo≈ºliwo≈õci jest przechowywanie odniesie≈Ñ do obiekt√≥w w innym (nadrzƒôdnym) obiekcie. W ≈õwiecie baz danych takie powiƒÖzanie nazywane by≈Çoby <strong style="color: #34495e;">'relacjƒÖ'</strong>.
            </p>
            <p style="margin: 10px 0 0 0; font-size: 1.1em; line-height: 1.6; color: #2c3e50;">
                Je≈õli naszym zamiarem jest przechowywanie tylko jednej instancji <code style="background-color: #ecf0f1; padding: 2px 6px; border-radius: 4px; color: #7f8c8d; font-weight: bold;">HelloMany</code>, mo≈ºemy oznaczyƒá jƒÖ jako <code style="background-color: #ecf0f1; padding: 2px 6px; border-radius: 4px; color: #7f8c8d; font-weight: bold;">singleton</code>.
            </p>
        </div>
    </div>
</div>

```python
@db0.memo(singleton = True)
class HelloMany:
    # przekazujemy listƒô imion i zostanie utworzonych wiele instancji HelloWorld, kt√≥re nastƒôpnie zostanƒÖ dodane do listy DB0 'hellos'
    def __init__(self, names):
        self.hellos = [HelloWorld(name) for name in names]
        
    def greet(self):
        for hw in self.hellos:
            hw.greet()
```

<div style="background-color: #ecf0f1; border: 1px solid #bdc3c7; border-radius: 6px; padding: 15px; margin: 15px 0;">
    <p style="margin: 0; display: flex; align-items: center; gap: 10px; color: #2c3e50; font-size: 1.1em;">
        <span style="font-size: 1.5em;">‚ö°</span>
        <strong>Wa≈ºne:&nbsp;</strong> Obiekt&nbsp;<strong>dbzero</strong>&nbsp; klasy oznaczonej jako &nbsp;<code style="background-color: #34495e; color: white; padding: 2px 6px; border-radius: 4px; font-weight: bold;">singleton</code>&nbsp; zostanie utworzony tylko raz.
    </p>
</div>

```python
many_hellos = HelloMany(["Andrzej", "Alicja", "Bartek", "Filip", "Robert"])
```

```python
many_hellos.greet()
```

<div style="background-color: #f8f9fa; border: 1px solid #95a5a6; border-radius: 6px; padding: 15px; margin: 15px 0;">
    <p style="margin: 0; display: flex; align-items: flex-start; gap: 10px; color: #2c3e50; font-size: 1.05em; line-height: 1.5;">
        <span style="font-size: 1.5em;">üîÑ</span>
        <span><strong>Ponowny dostƒôp:&nbsp;</strong> Po utworzeniu instancji obiektu mo≈ºna ponownie uzyskaƒá do niego dostƒôp (z innego programu lub po zamkniƒôciu tego programu) bez ≈ºadnych argument√≥w lub przy u≈ºyciu funkcji db0.fetch.</span>
    </p>
</div>

```python
mh = HelloMany()
mh.greet()
```

```python
db0.fetch(HelloMany)
```

<div style="background-color: #ecf0f1; border-left: 4px solid #34495e; padding: 15px; margin: 15px 0; border-radius: 5px;">
    <p style="margin: 0; font-size: 1.1em; color: #2c3e50;">
        üîç <strong>Obserwacja:</strong> Widzimy, ≈ºe <code style="background-color: #f39c12; color: white; padding: 2px 6px; border-radius: 4px; font-weight: bold;">many_hellos</code> i <code style="background-color: #f39c12; color: white; padding: 2px 6px; border-radius: 4px; font-weight: bold;">mh</code> to ten sam obiekt <strong>dbzero</strong> i nawet ten sam obiekt Python.
    </p>
</div>

```python
print(f"{id(many_hellos)} {id(mh)}")
```

```python
print(f"{db0.uuid(many_hellos)} {db0.uuid(mh)}")
```

                Podobnie jak w przypadku zwyk≈Çych obiekt√≥w Python - lista przechowuje <strong style="color: #34495e;">odniesienia</strong>, a nie pe≈Çne instancje. Listy <strong>dbzero</strong> mogƒÖ przechowywaƒá proste typy (liczby/ciƒÖgi znak√≥w), inne kolekcje (listy, s≈Çowniki, zbiory, krotki) lub odniesienia do innych instancji <code style="background-color: #ecf0f1; padding: 2px 6px; border-radius: 4px; color: #7f8c8d; font-weight: bold;">@memo</code>.
            </p>
        </div>
    </div>
</div>

```python
for obj in mh.hellos:
    print(db0.uuid(obj))
```
