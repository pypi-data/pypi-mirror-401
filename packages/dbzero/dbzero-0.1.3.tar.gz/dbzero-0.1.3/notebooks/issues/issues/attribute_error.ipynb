```python
from dataclasses import dataclass
import os
import dbzero as db0

DB0_DIR = os.path.join(os.getcwd(), "db0-test-data")
PREFIX = "my-test-prefix2"
```

```python
db0.init(DB0_DIR, config = {"autocommit": False})
db0.open(PREFIX)
```

```python
@db0.memo
# FIXME: log
# @dataclass
class Test:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def __iter__(self):
        return (x for x in [self.a, self.b, self.c])

    def __gt__(self, other):
        return self.a > other.a

    def times2(self):
        return self.a*2 
```

```python
obj1 = Test(1,2,3)
obj2 = Test(4,5,6)
db0.tags(obj1).add("tag1")
db0.tags(obj2).add("tag2")
db0.commit()
```

```python
list(db0.find(Test))
```

```python
list(db0.filter(lambda x: x.d + x.c == 4, db0.find("tag1")))
```

```python
obj1.times2()
```

```python
db0.close()
```

```python
@db0.memo
class Test2:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def __iter__(self):
        return (x for x in [self.a, self.b, self.c])

    def __gt__(self, other):
        return self.a > other.a

    def __lt__(self, other):
        return self.a < other.a

    def __mul__(self, other):
        return self.a * other.a

    # def __repr__(self):
    #     return repr("<Test 2>")

    def __eq__(self, other):
        return self.a == other.a

    def times2(self):
        return self.a*2 
```

```python
obj3 = Test2(1,2,3)
obj4 = Test2(4,5,6)
```
