```python
import os
import dbzero as db0

DB0_DIR = os.path.join(os.getcwd(), "db0-test-data")
PREFIX = "my-test-prefix"
```

```python
db0.init(DB0_DIR, config = {"autocommit": False})
db0.open(PREFIX)
```

```python
@db0.memo
class Test:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        db0.tags(self).add("tag1")
```

```python
obj1 = Test(1,2,3)
```
