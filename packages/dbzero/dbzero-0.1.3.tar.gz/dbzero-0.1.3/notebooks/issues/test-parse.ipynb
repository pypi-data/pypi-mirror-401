```python
import re
import os
```

```python
f = open("/src/dev/ops.txt", "r")
w = open("/src/dev/data.txt", "w")
```

```python
lines = f.readlines()
```

```python
lines[-1]
```

```python
i = 0
for l in lines:
    group = re.match(".*\(([0-9]*), ([0-9]*), ([0-9]*)\).*", l)
    page, state, storage = tuple([group[1], group[2], group[3]])
    w.write("{" + str(page) + ", " + str(state) + ", " + str(storage) + "},")
    i += 1
    if i % 10 == 0:
        w.write("\r\n")
```

```python
w.close()
```
