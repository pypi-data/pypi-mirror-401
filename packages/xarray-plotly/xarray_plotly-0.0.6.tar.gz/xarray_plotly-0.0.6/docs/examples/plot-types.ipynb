# Plot Types

Overview of all available plot types.

```python
import plotly.express as px
import xarray as xr

from xarray_plotly import config, xpx

config.notebook()
```

```python
# Sample data
df = px.data.stocks().set_index("date")
df.index = df.index.astype("datetime64[ns]")

stocks = xr.DataArray(
    df.values,
    dims=["date", "company"],
    coords={"date": df.index, "company": df.columns.tolist()},
    name="price",
)
```

## area

```python
xpx(stocks).area()
```

## bar

```python
xpx(stocks.isel(date=[0, 25, 50, 75, 100])).bar()
```

## box

```python
xpx(stocks).box()
```

## imshow

```python
xpx(stocks).imshow()
```

## line

```python
xpx(stocks).line()
```

## pie

```python
xpx(stocks.isel(date=-1)).pie()
```

## scatter

```python
xpx(stocks).scatter()
```
