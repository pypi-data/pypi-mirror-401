# The 'DatabaseDict' Class

```python
from powfacpy.applications.database import DatabaseDict
class_params = {
        "ElmLod": ["plini", "qlini"],
    }
obj_attr = pfdb.get_object_attributes(
    loads, 
    class_params, 
)
db_dict = DatabaseDict(obj_attr)
db_dict
```

```python
class_attr_values = {
    ("ElmLod","plini"): {
        "min": 0,     
        "max": lambda x: x.plini * 0.9,
    },
    ("ElmLod","qlini"): {
        "min": 0,     
        "max": 0.5,
    },
}

db_dict.set_selected_values(class_attr_values, set_in_pf=False)
db_dict
```
