```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Create widgets
amplitude_slider = widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='Amplitude:')
frequency_slider = widgets.FloatSlider(value=1.0, min=0.5, max=3.0, step=0.1, description='Frequency:')
update_button = widgets.Button(description="Update Plot")

controls = widgets.VBox([amplitude_slider, frequency_slider, update_button])

# Create an Output widget for the plot
plot_output = widgets.Output()

def update_plot(change=None):
    with plot_output:
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(10, 5))
        x = np.linspace(0, 2 * np.pi, 100)
        y = amplitude_slider.value * np.sin(frequency_slider.value * x)
        ax.plot(x, y)
        ax.set_ylim(-2, 2)
        display(fig)
        plt.close(fig)

update_button.on_click(update_plot)
amplitude_slider.observe(update_plot, names='value')
frequency_slider.observe(update_plot, names='value')

# Initial plot
update_plot()

# Arrange controls and plot side by side
dashboard = widgets.HBox([plot_output, controls])
display(dashboard)
```
