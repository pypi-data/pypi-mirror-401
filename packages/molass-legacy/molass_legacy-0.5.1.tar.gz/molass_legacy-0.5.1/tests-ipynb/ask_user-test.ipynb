```python
import ipywidgets as widgets
from IPython.display import display, clear_output

def ask_user(question, callback=None, output_widget=None):
    """
    Display a Yes/No dialog in a dedicated Output widget.
    - question: str, the question to display
    - callback: function, called with True (Yes) or False (No)
    - output_widget: ipywidgets.Output, where to display the dialog (optional)
    """
    if output_widget is None:
        output_widget = widgets.Output()
        display(output_widget)  # Only display if not already displayed

    with output_widget:
        clear_output(wait=True)
        label = widgets.Label(question)
        btn_yes = widgets.Button(description="Yes", button_style='success')
        btn_no = widgets.Button(description="No", button_style='danger')
        button_box = widgets.HBox([btn_yes, btn_no])
        display(widgets.VBox([label, button_box]))

        def on_yes(b):
            with output_widget:
                clear_output(wait=True)
                print("You selected: Yes")
            if callback:
                callback(True)

        def on_no(b):
            with output_widget:
                clear_output(wait=True)
                print("You selected: No")
            if callback:
                callback(False)

        btn_yes.on_click(on_yes)
        btn_no.on_click(on_no)

    return output_widget
```

```python
dialog_output = widgets.Output()
display(dialog_output)

def handle_response(answer):
    print("Callback received:", answer)
    # You can trigger further actions here
    if answer:
        print("Continuing operation...")

ask_user("Do you want to continue?", callback=handle_response, output_widget=dialog_output)
# returns immediately, response will be set later via callbac
```
