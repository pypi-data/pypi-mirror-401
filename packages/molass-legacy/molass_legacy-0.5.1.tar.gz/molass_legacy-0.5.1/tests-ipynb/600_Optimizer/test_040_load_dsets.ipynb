# Trimmned SD

```python
import os
import numpy as np
from molass_legacy._MOLASS.SerialSettings import set_setting
from molass_legacy.SecSaxs.DataTreatment import DataTreatment
from molass.Bridge.OptimizerInput import OptimizerInput
in_folder = r"C:\Users\takahashi\GitHub\molass-data\molass_data\SAMPLE4"
analysis_folder = r"C:\Users\takahashi\GitHub\molass-library\tests-ipynb\generic\temp_analysis"
optimizer_folder = os.path.join(analysis_folder, "optimized")
trimming_txt = os.path.join(optimizer_folder, "jobs", "000", "trimming.txt")

set_setting('analysis_folder', analysis_folder)
set_setting('optimizer_folder', optimizer_folder)

fullopt_input = OptimizerInput(in_folder=in_folder, trimming_txt=trimming_txt, legacy=True)
```

```python
dsets = fullopt_input.get_dsets()
```

```python
from molass_legacy.Optimizer.FuncImporter import import_objective_function
from molass_legacy.Baseline.BaselineUtils import create_xr_baseline_object

n_components = 4
class_code = 'G0346'
fullopt_class = import_objective_function(class_code)
uv_base_curve = fullopt_input.get_base_curve()      # uv_base_curve comes from FullOptInput.get_sd_from_folder()
xr_base_curve = create_xr_baseline_object()
qvector, wvector = fullopt_input.get_spectral_vectors()
optimizer = fullopt_class(dsets, n_components,
            uv_base_curve=uv_base_curve,
            xr_base_curve=xr_base_curve,
            qvector=qvector,   # trimmmed sd
            wvector=wvector,
            )
```

```python
from molass_legacy.Optimizer.Scripting import get_params
from molass_legacy.Optimizer.JobStatePlot import plot_objective_func
params = get_params(r"C:\Users\takahashi\GitHub\molass-library\tests-ipynb\generic\temp_analysis\optimized\jobs\000")
optimizer.prepare_for_optimization(params)
plot_objective_func(optimizer, params)
```
