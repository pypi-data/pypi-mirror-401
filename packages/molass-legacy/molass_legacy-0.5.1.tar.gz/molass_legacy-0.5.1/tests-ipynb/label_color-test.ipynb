```python
import ipywidgets as widgets
from IPython.display import display, HTML
import time

# 1. Define and display the label
label = widgets.Label(value="Status: Ready")
display(label)

# 2. Inject custom CSS (run once per notebook)
display(HTML("""
<style>
.status-label-green {
    color: green !important;
    font-weight: bold;
}
.status-label-red {
    color: red !important;
    font-weight: bold;
}
.status-label-yellow {
    color: orange !important;
    font-weight: bold;
}
</style>
"""))

# 3. Function to change label color
def set_label_color(label, color):
    # Remove all color classes
    for c in ["status-label-green", "status-label-red", "status-label-yellow"]:
        label.remove_class(c)
    # Add the selected color class
    label.add_class(f"status-label-{color}")

# 4. Unit-testable color changes
set_label_color(label, "green")
label.value = "Status: Success"
time.sleep(1)

set_label_color(label, "red")
label.value = "Status: Error"
time.sleep(1)

set_label_color(label, "yellow")
label.value = "Status: Warning"
```
