# Tutorial 1

## VirtualShip exercise
You can work on the exercise either individually or in pairs. You will work with the same person during the tutorials on November 17th as well. 
You will plan and execute a virtual oceanographic expedition using the VirtualShip software. Follow the steps below to complete the exercise.

### Form and register your duo (preferably with different backgrounds) through Teams
- Choose a month in the summer half year in which you want to do the expedition (i.e. April - October)
- Fill in the sheet in the Teams channel and remember your group number (e.g. GROUP1) and the year it is associated with (e.g. 1993)

### Check out your expeditions measurement locations
https://nioz.marinefacilitiesplanning.com/cruiselocationplanning#
- Save and upload the cruise track (.xlsx file)
- Select _Texel - Netherlands_ as the Port of Departure 
- Check out the measurement locations and depths, and travel time between stations
- Do a (rough) calculation on the time the CTD measurements will take at each station 
  - Assume 10 minutes for deployment and retrieval of the CTD at each station
  - Assume 1 m/s for the CTD to go down and up (i.e., for a station at 100 m depth, it will take approximately 200 seconds to go down and up)

### Open a terminal and prepare your expedition in VirtualShip:
```
cd data/virtualship_storage/
virtualship plan GROUP#
```
Replace # with your actual group number (e.g., GROUP1)

- Decide on the exact timing of your expedition with in the year and month registered though Teams
- Remember you have 9 days of ship time available, including travel time from and to Texel, Netherlands

### Follow the instructions in the terminal to set up your expedition
- From the Schedule Editor select Waypoints & Instrument Selection
- Click each waypoint to set the time (consider _transit times_ between locations and _time needed for measurements_) 
- Ensure that the CTD and CTD_BGC instruments are selected for use at all waypoints (they should be by default).
- Save your changes

### Fetch the data needed for your expedition
```
virtualship fetch GROUP#
```
Replace GROUP# with your actual expedition name (e.g., GROUP1)

- Provide your Copernicus username and password when prompted
- (Sign up for a Copernicus account if you don't have one yet: https://data.marine.copernicus.eu/register)
- Practice your patience - a key skill in oceanography! - data download may take some time ;-)

### Start your expedition
`virtualship run GROUP#`

### Verify your expedition was successful by checking the output directory for data files
- Navigate to the results directory:
`cd GROUP#/results`
- List the files in the results directory:
`ls`

- Hand in the filepath to your results via Brightspace before Wednesday 19-11-2025 13:00h:
