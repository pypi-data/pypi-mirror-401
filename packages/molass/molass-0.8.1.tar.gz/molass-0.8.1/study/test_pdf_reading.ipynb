```python
import sys
print(f"Python version: {sys.version}")
```

```python
# Test pypdf
try:
    import pypdf
    print(f"✓ pypdf version: {pypdf.__version__}")
    pypdf_available = True
except ImportError as e:
    print(f"✗ pypdf not available: {e}")
    pypdf_available = False
```

```python
# Test PyPDF2 as alternative
try:
    import PyPDF2
    print(f"✓ PyPDF2 version: {PyPDF2.__version__}")
    pypdf2_available = True
except ImportError as e:
    print(f"✗ PyPDF2 not available: {e}")
    pypdf2_available = False
```

```python
# If pypdf is available, test reading a PDF
if pypdf_available:
    from pypdf import PdfReader
    
    # Try to read one of your PDFs
    pdf_path = r"E:\GitHub\molass-library\study\2002, Francesco Dondi.pdf"
    
    try:
        reader = PdfReader(pdf_path)
        num_pages = len(reader.pages)
        print(f"\n✓ Successfully opened: {pdf_path}")
        print(f"  Number of pages: {num_pages}")
        
        # Extract text from first page as test
        if num_pages > 0:
            first_page_text = reader.pages[0].extract_text()
            print(f"  First 200 characters from page 1:")
            print(f"  {first_page_text[:200]}...")
    except Exception as e:
        print(f"✗ Error reading PDF: {e}")
elif pypdf2_available:
    # Use PyPDF2 as fallback
    from PyPDF2 import PdfReader
    print("Using PyPDF2 instead...")
    # Similar code as above
else:
    print("\n⚠ No PDF library available. Need to install pypdf:")
    print("  !pip install pypdf")
```
