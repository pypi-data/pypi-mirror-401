```python
from molass_data import SAMPLE1
from molass.DataObjects import SecSaxsData as SSD
ssd = SSD(SAMPLE1)
trimmed_ssd = ssd.trimmed_copy()
corrected_ssd = trimmed_ssd.corrected_copy()
corrected_ssd.plot_compact()
```

```python
decomposition = corrected_ssd.quick_decomposition(num_components=3)
pairedranges = decomposition.get_pairedranges()
decomposition.plot_components(pairedranges=pairedranges)
```

```python
start, stop = pairedranges[0].get_fromto_list()[0]
rslice = slice(start, stop + 1)
qv = corrected_ssd.xr.qv
D = corrected_ssd.xr.M[:,rslice]
E = corrected_ssd.xr.E[:,rslice]

from importlib import reload
import molass_legacy.BoundedLRF.BoundedLrfSolver
reload(molass_legacy.BoundedLRF.BoundedLrfSolver)
from molass_legacy.BoundedLRF.BoundedLrfSolver import BoundedLrfSolver
solver = BoundedLrfSolver(qv, D, E)
result = solver.solve(debug=True)
```

```python
import matplotlib.pyplot as plt
P_ = result[0]
P_.shape
plt.plot(qv, P_[:,0])
plt.yscale('log')
```
