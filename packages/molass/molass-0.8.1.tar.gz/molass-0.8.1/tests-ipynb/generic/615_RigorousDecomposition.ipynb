```python
from molass import get_version
assert get_version() >= '0.7.5', "this notebook requires molass v0.7.5 or higher."
from molass_data import SAMPLE4
from molass.DataObjects import SecSaxsData as SSD
```

```python
ssd = SSD(SAMPLE4)
trimmed_ssd = ssd.trimmed_copy()
corrected_ssd = trimmed_ssd.corrected_copy()
corrected_ssd.plot_compact();
```

```python
decomposition = corrected_ssd.quick_decomposition(proportions=[0.2, 0.5, 0.3])
decomposition.plot_components();
```

```python
rgcurve = corrected_ssd.xr.compute_rgcurve()
plot6 = decomposition.plot_components(title="EGH Decomposition", rgcurve=rgcurve)
```

```python
run_info = decomposition.optimize_rigorously(rgcurve=rgcurve, analysis_folder="temp_analysis", method='NS', niter=20, debug=True)
```

```python
current_decomposition = run_info.get_current_decomposition(wait_for_first_results=True, debug=True)
```

```python
current_decomposition.plot_components(title="Optimized EGH Decomposition", rgcurve=rgcurve)
```
